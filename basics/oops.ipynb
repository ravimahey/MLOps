```python
class BackAccount:
    def __init__(self,owner,balance):
        self.owner=owner
        self.balance=balance
    
    def deposit(self,amount):
        self.balance+=amount
        print(f"Deposit of {amount} is successful. New balance is {self.balance}")
    
    def withdraw(self,amount):
        if self.balance>=amount:
            self.balance-=amount
            print(f"Withdrawal of {amount} is successful. New balance is {self.balance}")
        else:
            print(f"Insufficient balance. Current balance is {self.balance}")
    
    def __str__(self):
        return f"Account holder: {self.owner}\nBalance: {self.balance}"
    
account = BackAccount("John Doe", 1000)
account.deposit(500)
account.withdraw(200)
account.withdraw(22200)
print(account)
```

```python
# Ploymorphism

class Dog:

    def __init__(self,name,age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name} is {self.age} years old"
    
    def bark(self):
        print(f"{self.name} is barking on owner")

class GermanShepherd(Dog):
    def __init__(self,name,age,color):
        super().__init__(name,age)
        self.color = color
    
    def bark(self):
        print(f"{self.name} is barking and barking")

class GoldenRetriever(Dog):
    def __init__(self,name,age,color):
        super().__init__(name,age)
        self.color = color

    def bark(self):
        print(f"{self.name} is barking")

common_dog= Dog("Libra",3)
libra_dog= GermanShepherd("Libra",3,"White")
rex_dog = GoldenRetriever("Rex",4,"Golden")

common_dog.bark()
libra_dog.bark()
rex_dog.bark()
```

```python
#  Interface and Abstract base Classes

from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def make_sound(self):
        pass
    @abstractmethod
    def move(self):
        pass
    @abstractmethod
    def eat(self):
        pass
    @abstractmethod
    def sleep(self):
        pass

class Dog(Animal):
    def make_sound(self):
        print("Woof Woof")
    def move(self):
        print("Walking")
    def eat(self):
        print("Eating")
    def sleep(self):
        print("Sleeping")

class Cat(Animal):
    def make_sound(self):
        print("Meow Meow")
    def move(self):
        print("Running")
    def eat(self):
        print("Eating")
    def sleep(self):
        print("Sleeping")


dog = Dog()
dog.make_sound()
dog.move()
dog.eat()

cat = Cat()
cat.make_sound()
cat.move()
```

```python
# Encapsulation and Abstraction
# public, protected, private

from abc import ABC, abstractmethod
class Person:
    def __init__(self,name,age): 
        self.__name = name
        self.__age = age

    def get_name(self):
        return self.__name
    
    def get_age(self):
        return self.__age
    
    def __set_name(self,name):
        self.__name = name

    def set_age(self,age):
        self.__age = age

    def __str__(self):
        return f"{self.__name} is {self.__age} years old"
    
def get_name(person):
    return person.name

person1 = Person("John Doe", 30)
print(person1.get_name())
print(person1.get_age())
person1.set_name("Jane Doe")
person1.set_age(25)
print(person1)
print(person1)

# print(get_name(person1))
print(dir(person1))

```

```python
# Abstraction 

from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass
    @abstractmethod
    def stop(self):
        pass

class Car(Vehicle):
    def start(self):
        print("Starting the car")
    def stop(self):
        print("Stopping the car")

car = Car()
car.start()
car.stop()
```

```python
# Custom Exception

class Errors(Exception):
    pass

class DobException(Errors):
    pass


class AgeException(Errors):
    pass


class NameException(Errors):
    pass


def validate(name, age, dob):
    try:
        if name == "":
            raise NameException("Name cannot be empty")
        if age < 0:
            raise AgeException("Age cannot be negative")
        if dob == "":
            raise DobException("Date of birth cannot be empty")
    except NameException as e:
        print(e)
    except AgeException as e:
        print(e)
    except DobException as e:
        print(e)

validate("adsfsad", 30, "1990-01-01")
```

```python
# Operator Overloading

class Dog:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name} is {self.age} years old"
    
    def __add__(self,other):
        return self.age + other.age
    def __mul__(self, other):
        return self.age * other.age
    

libra_dog= Dog("Libra",3)
rex_dog = Dog("Rex",4)

print(libra_dog + rex_dog)
print(libra_dog * rex_dog)
```

```python
# Generators

def square(n):
    for i in range(n):  
        yield i**2

print(square(5))

for i in square(5):
    print(i, end=" ")



```

```python
# Load Large Files 

import os
def load_large_file(file_path):
    with open(file_path, "r") as file:
        for line in file:
            yield line

file_path=os.path.join('testing',"logs.txt")

for line in load_large_file(file_path):
    print(line)
```
