```python
# Vectorized operation
arr1=np.array([1,2,3,4,5])
arr2=np.array([2,3,4,5,6])
print(arr1+arr2)

# Substraction
print(arr2-arr1)

print(arr1*arr2)

print(arr1/arr2)

print(arr1**2)

print(arr1>2)

print(arr1[1:4])
```

```python
# Universal Functions
arr=np.array([1,2,3,4,5])

print(np.sqrt(arr))

print(np.exp(arr))

print(np.sign(arr))

print(np.log(arr))

print(np.log10(arr))

print(np.sin(arr))
```

```python
# Array Slicing and Indexing
arr=np.array([[1,2,3,4,5],[1,2,3,3,4],[1,2,3,4,5]])

print(arr[0][0])

print(arr[0,0])

print(arr[1:3,2:])
```
