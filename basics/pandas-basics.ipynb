```python
csv_file = 'data\\sales_data_sample.csv'
sales_df = pd.read_csv(csv_file)
sales_df.head(5)
```

```python
groupped_sales = sales_df.groupby('COUNTRY')['SALES'].agg(['sum','mean', 'max', 'min'])
groupped_sales
```

```python
### Merging and Joining DataFrames

df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35]
})

df2 = pd.DataFrame({
    'ID': [2, 3, 4],
    'City': ['New York', 'London', 'Paris'],
    'Country': ['USA', 'UK', 'France']
})

merged_df = pd.merge(df1, df2, on='ID', how='outer')
print(merged_df)
```
