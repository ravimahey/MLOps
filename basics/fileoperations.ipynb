```python
# Read a file
# print("Hello workd")
with open("testing\\test.txt", 'r') as f:
    content = f.read()
    print(content)
    # print(f.read())
    for line in f:
        print(line.strip())
    
    f.close()

with open("testing\\source.txt", 'w') as f:
    f.write("Hello world\n")

with open("testing\\source.txt", 'a') as f:
    f.write("Hello world\n")
    f.write("Hello world\n")
    f.write("Hello world\n")


lines = ["Hello world", "Hello world", "Hello world"]
with open("testing\\source.txt", 'a') as f:
    f.writelines(lines)


# Binary files
data = b'\x01\x02\x03'
with open("testing\\binary.bin", 'wb') as f:
    f.write(data)

with open("testing\\binary.bin", 'rb') as f:
    data = f.read()
    print(data)

with open("testing\\source.txt", 'r') as f:
    content = f.read()

with open("testing\\destination.txt", 'w') as f:
    f.write(content)

# Write and reading
with open("testing\\test.txt", 'w+') as f:
    f.write("name,age,city\n")
    f.seek(10)
    f.write("John,30,New York\n")
    f.seek(20)
    f.write("Jane,25,Los Angeles\n")
    content = f.read()
    print(content)

with open("testing\\example.txt", "r") as f:
    f.seek(5)  # move to 5th byte
    data = f.read()
    print(data)

with open("testing\\example.txt", "r") as f:
    print(f.read(5))   # Hello
    f.seek(0)          # go back to start
    print(f.read(5))   # Hello again
```
